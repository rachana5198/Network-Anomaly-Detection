```python
 X_train, X_test,y_train,y_test = train_test_split(df,labels, test_size=0.1,random_state=42)
```

```python
#Training using KNN (Supervised Learning)
knn = KNeighborsClassifier(n_neighbors = 5, algorithm = 'ball_tree', leaf_size=500)
t0 = time()
knn.fit(X_train,y_train)
tt = time()-t0
print("Classifier trained in {} seconds".format(round(tt,3)))
```

```python
#Prediction using KNN
t0 = time()
pred = knn.predict(X_test)
tt = time() - t0
print("Predicted in {} seconds".format(round(tt,3)))
```

```python
#Calculating accuracy of KNN

acc = accuracy_score(pred, y_test)
print("Accuracy of KNN is {}.".format(round(acc,4)))
```

```python
#Unsupervised learning (K means clustering)

from sklearn.cluster import KMeans
k = 30
km = KMeans(n_clusters = k)
```

```python
#Training into clusters using k-means 
t0 = time()
km.fit(X_test)
tt = time()-t0
print("Clustered in {} seconds".format(round(tt,3)))
```

```python
#checking the cluster sizes
pd.Series(km.labels_).value_counts()
```

```python
label = df1['Attack_names']
label_names = list(map(
    lambda x: pd.Series([label[i] for i in range(len(km.labels_)) if km.labels_[i]==x]), 
    range(k)))
```

```python
#Print labels for each cluster.
for i in range(k):
    print("Cluster {} labels:".format(i))
    print(label_names[i].value_counts())
    print()
```

```python
#Predicting using k-means for test data
t0 = time()
pred = km.predict(X_test)
tt = time() - t0
print("Assigned clusters in {} seconds".format(round(tt,3)))
```
