```python
 X_train, X_test,y_train,y_test = train_test_split(df,labels, test_size=0.1,random_state=42)
```

```python
#Training using KNN (Supervised Learning)
knn = KNeighborsClassifier(n_neighbors = 5, algorithm = 'ball_tree', leaf_size=500)
t0 = time()
knn.fit(X_train,y_train)
tt = time()-t0
print("Classifier trained in {} seconds".format(round(tt,3)))
```

```python
#Prediction using KNN
t0 = time()
pred = knn.predict(X_test)
tt = time() - t0
print("Predicted in {} seconds".format(round(tt,3)))
```

```python
#Calculating accuracy of KNN

acc = accuracy_score(pred, y_test)
print("Accuracy of KNN is {}.".format(round(acc,4)))
```

```python
#ML  PIPELINE
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
pipeline = make_pipeline(StandardScaler(),KNeighborsClassifier(n_neighbors = 5, algorithm = 'ball_tree', leaf_size=500) )
```

```python
model = pipeline.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
```

```python
from sklearn import metrics
metrics.classification_report(y_test, predicted)
```
